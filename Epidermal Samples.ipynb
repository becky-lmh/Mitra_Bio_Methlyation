```python
# Create sample headings
OE = ["OE"]*10
OP = ["OP"]*10
YE = ["YE"]*9
YP = ["YP"]*9
sample_list = OE+OP+YE+YP
```

```python
# Clean up the empty cells
df1 = df.replace(r'^\s*$', np.nan, regex=True)
df2 = df1.astype(float)
#print(df2.dtypes)
```

```python
# Create boxplot
df_box=df2
df_box.columns = sample_list
#df_box.columns = ["a","b"]
fig1, ax1 = plt.subplots(figsize=(12,7))
ax1.set_title("Methylation Values for Epidermal Samples")
boxplot = df_box.boxplot()

plt.show()
```

```python
# PCA Analysis

# create dataframe for PCA
df_PCA=df2.T

#targets = pd.DataFrame(["a","b"], columns=["target"])
targets = pd.DataFrame(sample_list, columns=["target"])
#df.head()
#targets.head()

#remove NAN and transpose
df_PCA1 = df_PCA.dropna(axis=1, how='any')
print("%d rows with missing values were removed." % (len(df_PCA.columns) - len(df_PCA1.columns)))
#print (df2.head())

# Standardising values
x = df_PCA1.values
x = StandardScaler().fit_transform(x)

pca = PCA(n_components = 2, random_state = 0)
principalComponents = pca.fit_transform(x)
pca_names = [str(i)+str(j) for i,j in zip(['PC']*2,range(1,3))]
principalDf = pd.DataFrame(data = principalComponents, columns = pca_names)

finalDf = pd.concat([principalDf, targets], axis = 1)
finalDf.head()
```

```python
#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
fig4, ax4 = plt.subplots(figsize=(7,3))
#colors = ["red","green","blue","yellow","pink","black",
            #"orange","purple","beige","brown","gray",
            #"cyan","magenta"])
plt.scatter(principalComponents[:10, 0], principalComponents[:10, 1],
            edgecolor='none', alpha=0.5, s=200, 
            color='blue', label='OE')
plt.scatter(principalComponents[10:20, 0], principalComponents[10:20, 1],
            edgecolor='none', alpha=0.5, s=200,
            color='yellow', label='OP')
plt.scatter(principalComponents[20:29, 0], principalComponents[20:29, 1],
            edgecolor='none', alpha=0.5, s=200,
            color='green', label='YE')
plt.scatter(principalComponents[29:, 0], principalComponents[29:, 1],
            edgecolor='none', alpha=0.5, s=200,
            color='red', label='YP')

plt.legend(loc='best')
plt.title('2 Component PCA - Epidermal Samples', fontsize = 20)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.show()
```

```python
pca_vars = pca.explained_variance_ratio_
for n,v in zip(pca_names, pca_vars):
    print ("Variance explained by %s : %.4f%%" % (n, v*100))
```

```python
pca2 = PCA().fit(x)
plt.plot(np.cumsum(pca2.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
# create 3D PCA analysis

# Run The PCA
pca3d = PCA(n_components = 3, random_state = 0)
df3d = pca3d.fit_transform(x)
 
# Store results of PCA in a data frame
result=pd.DataFrame(df3d, columns=['PCA%i' % i for i in range(3)])
```

```python
# Plot initialisation
fig3 = plt.figure()
ax3 = fig3.add_subplot(111, projection='3d')
ax3.scatter(result['PCA0'].iloc[:10], result['PCA1'].iloc[:10], result['PCA2'].iloc[:10],
            edgecolor='none', alpha=0.5, s=50, 
            color='blue', label='OE')
ax3.scatter(result['PCA0'].iloc[10:20], result['PCA1'].iloc[10:20], result['PCA2'].iloc[10:20],
            edgecolor='none', alpha=0.5, s=50,
            color='yellow', label='OP')
ax3.scatter(result['PCA0'].iloc[20:29], result['PCA1'].iloc[20:29], result['PCA2'].iloc[20:29],
            edgecolor='none', alpha=0.5, s=50,
            color='green', label='YE')
ax3.scatter(result['PCA0'].iloc[29:], result['PCA1'].iloc[29:], result['PCA2'].iloc[29:],
            edgecolor='none', alpha=0.5, s=50,
            color='red', label='YP')
 
# make simple, bare axis lines through space:
xAxisLine = ((min(result['PCA0']), max(result['PCA0'])), (0, 0), (0,0))
ax3.plot(xAxisLine[0], xAxisLine[1], xAxisLine[2], 'r')
yAxisLine = ((0, 0), (min(result['PCA1']), max(result['PCA1'])), (0,0))
ax3.plot(yAxisLine[0], yAxisLine[1], yAxisLine[2], 'r')
zAxisLine = ((0, 0), (0,0), (min(result['PCA2']), max(result['PCA2'])))
ax3.plot(zAxisLine[0], zAxisLine[1], zAxisLine[2], 'r')
 
# label the axes
ax3.set_xlabel("PC1")
ax3.set_ylabel("PC2")
ax3.set_zlabel("PC3")
ax3.set_title("PCA in 3D - Epidermal Samples")
ax3.legend()
plt.show()
```
